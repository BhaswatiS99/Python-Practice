# 9) Aggregation & Group By

```python
df['Month'].value_counts()                  #.value_count() gives frequency of values
```

```python
df[df['Month']==1].value_counts()                #counts the values in Jan month, 1,2,3.. - Jan, Feb, March etc..
```

```python
#aggregation based on group by

df.groupby('Month')['Salary'].sum()                 #returns total salary in every month
```

```python
df.groupby('Month')['Salary'].mean()                 #returns average salary in every month
```

```python
df.groupby('Month').agg({'Salary' : 'mean', 'Name' : 'count'})          #65k is avg salary of 3 people, 20k is avg of 1 person
```

# 10) Concatenate & Merge DataFrame(JOINS)

```python
df1 = pd.DataFrame({'ID' : [1,2,3], 'Name' : ['A','B','C']})
print(df1)
```

```python
df2 = pd.DataFrame({'ID' : [1,2,2,4], 'Score' : [88,96,77,79]})
print(df2)
```

```python
pd.concat([df1,df2], axis=0)           #vertically joining two tables/row level/top on top
```

```python
pd.concat([df1,df2], axis=1)                  #horizontally joining two tables/column level/side on side
```

```python
#merge = join

pd.merge(df1,df2, how='inner', on= 'ID')             #inner join
```

```python
pd.merge(df1,df2, how='outer', on= 'ID')             #outer join
```

```python
pd.merge(df1,df2, how='left', on= 'ID')              #left join
```

```python
pd.merge(df1,df2, how='inner', left_on= 'ID', right_on= 'ID')
```
