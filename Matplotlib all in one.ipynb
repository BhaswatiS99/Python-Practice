```python
import matplotlib.pyplot as plt

#pyplot is a collection of functions that make matplotlib work as MATLAB
```

# 1) Create first chart

```python
# Data

x = [1,2,3,4,5]
y = [10,20,25,35,45]

plt.plot(x,y)                 #line chart
plt.show                      #display chart in jupyter notebook
```

# 2) Customized charts

```python
plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.show()                  #google matplotlib linechart parameters

plt.figure(figsize= (6,4))          #figure size
plt.show()
```

```python
plt.figure(figsize= (4,2))          #figure size

plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.show()
```

```python
# add heading

plt.figure(figsize= (4,2))             #figure size

plt.plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12)

plt.title("Figure 1")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.show()
```

# 3) Advance Line Chart- multiple lines & legends

```python
# Data

x = [1,2,3,4,5]
y1 = [10,20,25,35,45]
y2 = [20,30,35,45,55]

plt.plot(x, y1, label= "Sales 2024")          #plotting y1 data
plt.plot(x, y2, label= "Sales 2025")          #plotting y2 data

plt.title("YoY Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.legend()
plt.show()
```

# 4) Bar Chart

```python
#Shows categorical data not continuous

x = [1,2,3,4,5]
y = [10,20,55,35,45]

plt.bar(x,y)
plt.title("Bar chart example")
plt.show()
```

# 5) Histogram

```python
#Used for distribution analysis
#Shows continuous data 
#data

import random
data = [random.randint(1,20) for _ in range(100)]

plt.hist(data, bins= 20)
plt.title("Histogram Example")
plt.show()
```

# 6) Pie Chart

```python
#Shows relationship of whole data - part-to-whole

categories= ['A', 'B', 'C', 'D', 'E']
sales= [10,20,55,35,45]

plt.pie(sales, labels = categories,autopct= '%1.1f%%', startangle= 90)
plt.title("Pie Chart Example")
plt.show()
```

# 7) Scatter Plot 

```python
# Used to find relationships between variables

y1 = [10,20,25,35,45]
y2 = [20,30,35,45,55]

plt.scatter(y1,y2)
plt.title("Scatter Plot Example")
plt.show()
```

# 8) Subplots

```python
# Used to show multiple data in one figure

#data1 - bar chart
categories= ['M', 'T', 'W', 'T', 'F']
sales= [10,20,55,35,45]

#data2 - scatter plot
y1 = [10,20,25,35,45]
y2 = [20,30,35,45,55]

plt.figure(figsize= (8,4))

#1st plot - bar chart
plt.subplot(1,2,1)                     #row,column,position
plt.bar(categories, sales)
plt.title("Dailyly sales")
plt.xlabel("Week days")
plt.ylabel("Sales")

# 2nd plot - scatter plot
plt.subplot(1,2,2)
plt.scatter(y1,y2)
plt.title("User example") 
plt.xlabel("User1")
plt.ylabel("User2")
plt.show()
```

# 9) Matplotlib with Pandas - real data 

```python
# create dataframe
```

```python
import pandas as pd

#data
data = {
    'Month' : ['Jan', 'Feb', 'Mar', 'Apr'],
    'Sales' : [12000, 11000, 13000, 25000]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.bar(df['Month'], df['Sales'])              #x,y values as dataframes Month and Sales
plt.title("Matplotlib with Pandas")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

# 10) Save charts

```python
plt.bar(df['Month'], df['Sales'])              #x,y values as dataframes Month and Sales
plt.title("Matplotlib with Pandas")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
plt.savefig("Monthly_sales_from_df.png")
```
