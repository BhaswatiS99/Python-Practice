```python
#Q2) Assign radius = 7 and pi = 3.14159. Calculate and print the area and circumference using those variables.
```

```python
r = 7
pi = 3.14159
area = pi*r**2
circumference = 2*pi*r
print("Area of the circle: " , area)
print("Circumference of the circle: " , circumference)
```
