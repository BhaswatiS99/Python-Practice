```python
#Q13) Start with value = "55". Convert to int → multiply by 2.5 (becomes float) → convert to int → convert to string → concatenate " kg". Print the result at every step with the type.
```

```python
value = "55"
print(type(value))
value1 = int(value)
print(type(value1))
num = value1 * 2.5
print(num)
print(type(num))
num1 = int(num)
print(type(num1))
num2 = str(num)
print(type(num2))
concat = num2 + "kg"
print(concat)
print(type(concat))
```

```python
#Q14) Assign x = 7 (int) and y = 3.0 (float). Perform all arithmetic operations and print each result's value AND type. Note which operations return int vs float.
```
